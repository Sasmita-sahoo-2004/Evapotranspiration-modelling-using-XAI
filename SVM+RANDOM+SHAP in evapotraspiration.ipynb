```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv("C:\\Users\\HP\\OneDrive\\Desktop\\Internship Domain\\Plant.csv")

if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])  # Convert to datetime
    df.drop(columns=['date'], inplace=True)  # Drop the date column

# Here we are ensuring that all remaining columns are numeric
df = df.apply(pd.to_numeric, errors='coerce')

# if any missing value then fill it by mean 
df.fillna(df.mean(), inplace=True)
X = df.drop(columns=['pot_evapot']).values  # Convert to NumPy array
y = df['pot_evapot'].values

time_steps = 10
X_sequences = []

for i in range(len(X) - time_steps):
    X_sequences.append(X[i:i + time_steps])
X_sequences = np.array(X_sequences)

# Adjusting target to match the number of sequences
y_sequences = y[time_steps:]

# Spliting the dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(X_sequences, y_sequences, test_size=0.2, random_state=42)


input_layer = Input(shape=(time_steps, X.shape[1]))  
lstm_layer1 = LSTM(100, return_sequences=True)(input_layer)
dropout_layer1 = Dropout(0.2)(lstm_layer1)
lstm_layer2 = LSTM(100, return_sequences=True)(dropout_layer1)
dropout_layer2 = Dropout(0.2)(lstm_layer2)
lstm_layer3 = LSTM(100)(dropout_layer2)
output_layer = Dense(1)(lstm_layer3)

model = Model(inputs=input_layer, outputs=output_layer)
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mean_squared_error')


history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)

# Evaluating the model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error:', mse)
print('R² Score:', r2)

plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Plot Actual vs Predicted Values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt
import shap

df = pd.read_csv("C:\\Users\\HP\\OneDrive\\Desktop\\Internship Domain\\Plant.csv")

features = df[['max_wind_gust', 'avg_wind_speed', 'sol_rad',
                'max_air_temp', 'min_air_temp', 'avg_air_temp',
                'max_rel_hum', 'min_rel_hum', 'avg_rel_hum',
                'avg_dewpt_temp', 'precip']]
target = df['pot_evapot']

scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)

X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=0)

X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

history = model.fit(X_train, y_train, epochs=50, batch_size=32,
                    validation_data=(X_test, y_test), verbose=2, shuffle=False)

y_pred = model.predict(X_test)

mse = np.mean((y_test - y_pred.flatten())**2)
print(f'Mean Squared Error: {mse}')

def wrapped_model(X):
    X_reshaped = X.reshape((X.shape[0], 1, X.shape[1]))  # Reshape for LSTM
    return model.predict(X_reshaped).flatten()

X_test_sample = X_test[np.random.choice(X_test.shape[0], size=30, replace=False)]

explainer = shap.KernelExplainer(wrapped_model, shap.sample(X_train.reshape(X_train.shape[0], X_train.shape[2]), 100))
shap_values = explainer.shap_values(X_test_sample.reshape(X_test_sample.shape[0], X_test_sample.shape[2]), nsamples=100)

# SHAP summary plot
shap.summary_plot(shap_values, X_test_sample.reshape(X_test_sample.shape[0], X_test_sample.shape[2]), feature_names=features.columns)

# SHAP dependence plot 
shap.dependence_plot("avg_wind_speed", shap_values, X_test_sample.reshape(X_test_sample.shape[0], X_test_sample.shape[2]), feature_names=features.columns)

# SHAP feature importance plot
shap.summary_plot(shap_values, X_test_sample.reshape(X_test_sample.shape[0], X_test_sample.shape[2]), feature_names=features.columns, plot_type='bar')

# SHAP decision plot
shap.decision_plot(explainer.expected_value, np.array(shap_values)[:10], X_test_sample[:10].reshape(10, X_test_sample.shape[2]), feature_names=features.columns)

# Initialize JS visualization 
shap.initjs()

# SHAP force plot 
shap.force_plot(explainer.expected_value, shap_values[0][0], X_test_sample[0].reshape(1, X_test_sample.shape[2]), feature_names=features.columns)
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# Reading the dataset
df = pd.read_csv("C:\\Users\\HP\\OneDrive\\Desktop\\Internship Domain\\Plant.csv")
print(df.head())

train_dates = pd.to_datetime(df['date'])
print(train_dates.tail(15)) # Check the last few dates

# Columns for training
cols = list(df)[1:12]
print(cols) 

# New dataframe with only training data - excluding the date column
df_for_training = df[cols].astype(float)

# Normalizing our dataset
scaler = StandardScaler()
scaler = scaler.fit(df_for_training)
df_for_training_scaled = scaler.transform(df_for_training)

# Preparing data for LSTM
trainX = []
trainY = []

n_future = 1   # Number of days we want to look into the future based on the past days.
n_past = 14  # Number of past days we want to use to predict the future.

for i in range(n_past, len(df_for_training_scaled) - n_future + 1):
    trainX.append(df_for_training_scaled[i - n_past:i, 0:df_for_training.shape[1]])
    trainY.append(df_for_training_scaled[i + n_future - 1:i + n_future, 0])

trainX, trainY = np.array(trainX), np.array(trainY)

print('trainX shape == {}.'.format(trainX.shape))
print('trainY shape == {}.'.format(trainY.shape))

# Creating the LSTM model
model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
model.add(LSTM(32, activation='relu', return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(trainY.shape[1]))

model.compile(optimizer='adam', loss='mse')
model.summary()

history = model.fit(trainX, trainY, epochs=50, batch_size=16, validation_split=0.1, verbose=1)

# Plot the training and validation loss
plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.legend()
plt.show()

# Predicting...
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay
us_bd = CustomBusinessDay(calendar=USFederalHolidayCalendar())

n_days_for_prediction = 15  # Predict the past 15 days
predict_period_dates = pd.date_range(list(train_dates)[-n_past], periods=n_days_for_prediction, freq=us_bd).tolist()
print(predict_period_dates)

# Make prediction
prediction = model.predict(trainX[-n_days_for_prediction:]) # shape = (n, 1) where n is the n_days_for_prediction

# Perform inverse transformation to rescale back to original range
prediction_copies = np.repeat(prediction, df_for_training.shape[1], axis=-1)
y_pred_future = scaler.inverse_transform(prediction_copies)[:,0]

# Create forecast dataframe
forecast_dates = [time_i.date() for time_i in predict_period_dates]
df_forecast = pd.DataFrame({'date': np.array(forecast_dates), 'pot_evapot': y_pred_future})
df_forecast['date'] = pd.to_datetime(df_forecast['date'])

# Loading our original dataset and filter
original = df[['date', 'pot_evapot']]
original['date'] = pd.to_datetime(original['date'])
original = original.loc[original['date'] >= '2020-05-01']


plt.figure(figsize=(12, 6))
sns.lineplot(data=original, x='date', y='pot_evapot', label='Original')
sns.lineplot(data=df_forecast, x='date', y='pot_evapot', label='Forecast')
plt.title('Original vs Forecasted Pot_Evapot')
plt.xlabel('Date')
plt.ylabel('Potential Evapotranspiration')
plt.legend()
plt.show()

# Evaluating the model
mse = model.evaluate(trainX, trainY)
print('Mean Squared Error:', mse)
```
